###2. Data Visualization with Matplotlib/Seaborn

```python
#Create a histogram to visualize the distribution of a numerical variable:
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df['Price_USD'], kde = True)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Plot a bar chart to show the frequency of categories in a categorical variable:
sns.countplot(x= 'Category', data= df)
plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```

```python
#Create a scatter plot to visualize the relationship between two numerical variables:
sns.scatterplot(x= 'Price_USD' , y = 'Weight_pounds', data= df)
plt.title('Price vs Weight_pounds')
plt.xlabel('Price')
plt.ylabel('Weight_pounds')
plt.show()
```

###3. Data Aggregation and Grouping

```python
#Group data by a categorical variable and calculate statistics:
grouped_data = df.groupby('Category')['Price_USD'].mean()
print(grouped_data)
```

```python
grouped_data_1 = df.groupby('Category').agg({
    'Price_USD': ['mean', 'median', 'sum']
})
grouped_data_1
```
